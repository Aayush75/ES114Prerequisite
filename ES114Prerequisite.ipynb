```python
# Problem 1
l=[i for i in range(1,21)]
even_sum=0
odd_product=1
total=0
for i in l:
    if(i&1):                                   # if x&1 is not equal to 0, it means that x is an odd number
        odd_product*=i
    else:
        even_sum+=i
    total+=i
mean=total/len(l)
if len(l)&1:
    median=l[len(l)//2]                        # Finding the middle element in the sorted list if the length is an odd number
else:
    median=(l[len(l)//2-1]+l[len(l)//2])/2     # Mean of the two middle elements otherwise
sq_sum=0
for i in l:
    sq_sum+=(i-mean)**2
stdev=(sq_sum/len(l))**0.5                     # Using standard deviation formula
print(f"Sum of All Even Numbers in The List: {even_sum}")
print(f"Product of All Odd Numbers in The List: {odd_product}")
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Standard Deviation: {stdev}")
```

```python
# Problem 3
import matplotlib.pyplot as plt
n=int(input())
def fib(n):
    l=[0 for i in range(n+1)]
    l[1]=1
    for i in range(2,n+1):
        l[i]=l[i-1]+l[i-2]
    return l
plt.plot(fib(n),marker='.',linestyle='-',color='r',label='Fibonacci Numbers')
plt.title("Line Graph of Fibonacci Numbers")
plt.xlabel(r"$n$")
plt.ylabel(r"$F(n)$")          # Using LaTeX to make expressions look nice
plt.legend()
plt.show()
```

```python
# Problem 10
def make_dict(s):
    d={}
    for i in range(0,26):
        d[chr(ord("a")+i)]=0          # Initializing a dictionary with keys as all lowercase characters, and assigning the value 0 to them
    for c in s:
        if c.isalpha():
            c=c.lower()
            d[c]+=1
    return d
dict=make_dict(input())
x=list(dict.keys())
y=[dict[i] for i in x]
plt.figure(figsize=(12,7))
plt.bar(x,y,color='r')
plt.title("Frequency of Lowercase Characters in Input",fontsize=14)
plt.xlabel("Characters",fontsize=12)
plt.ylabel("Frequency",fontsize=12)
plt.show()
```

```python
# Problem 16
from matplotlib_venn import venn2
def intersection(a,b):
    c=[]
    vis={}                                 # Using hashmaps everywhere to avoid using 'in' each time, making all of the three functions O(|a|+|b|)
    for i in a:
        vis[i]=True
    for i in b:
        try:
            vis[i]
            c.append(i)
        except:
            continue
    return c

def union(a,b):
    c=[]
    vis={}
    for i in a:
        vis[i]=True
        c.append(i)
    for i in b:
        try:
            vis[i]
            continue
        except:
            vis[i]=True
            c.append(i)
    return c

def diff(a,b):
    c=[]
    vis={}
    for i in a:
        vis[i]=True
    for i in b:
        try:
            if vis[i]:
                vis[i]=False
        except:
            vis[i]=True
    for i in vis:
        if vis[i]:
            c.append(i)
    return c
    
a=set(map(int,input().split()))
b=set(map(int,input().split()))
venn=venn2([a,b],("A","B"))
venn.get_label_by_id("10").set_text("A")
venn.get_label_by_id("01").set_text("B")
venn.get_label_by_id("11").set_text("Intersection")
plt.title("Venn Diagram of A and B",fontsize=14)
print(f"A Intersection B = {intersection(a,b)}\nA Union B = {union(a,b)}\nSymmetric Difference = {diff(a,b)}")
```

```python
# Problem 21
l=list(map(int,input().split()))
total=0
freq={}
for i in l:
    total+=i
    freq[i]=freq.get(i,0)+1
mean=total/len(l)
x=sorted(l)
if len(x)&1:
    median=x[len(x)//2]
else:
    median=(x[len(x)//2-1]+x[len(x)//2])/2
sq_sum=0
for i in l:
    sq_sum+=(i-mean)**2
modes=[]
mx=0
for i in freq:
    if freq[i]>mx:
        modes.clear()
        modes.append(i)
        mx=freq[i]
    elif freq[i]==mx:
        modes.append(i)
stdev=(sq_sum/len(l))**0.5
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode(s): {", ".join([str(i) for i in sorted(modes)])}")
print(f"Standard Deviation: {stdev}")
plt.figure(figsize=(14,6))
plt.boxplot(l,patch_artist=True,boxprops=dict(facecolor="b"));
plt.title("Boxplot of Values in the List",fontsize=14)
plt.ylabel("Value",fontsize=12)
plt.show()
```

```python
# Problem 22
import numpy as np
def compress(x):                        # This function always works since lim x->INF f(x)=1, lim x->-INF f(x)=0,
    return 1/(1+np.exp(-x))             # f(x) is continuous on R, and it always lies in (0,1)
x=[(0.1)*i for i in range(-250,250)]
y=[compress(i) for i in x]
plt.figure(figsize=(14,6))
plt.title(r"Plot of $x$ vs $compress(x)$ for $-50 \leq x \leq 50$ with a step of 0.1",fontsize=14)
plt.xlabel(r"$x$",fontsize=12)
plt.ylabel(r"$compress(x)$",fontsize=12)
plt.plot(x,y,marker=".",color="r",linestyle="-");
```
